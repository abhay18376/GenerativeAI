```python
#Import the Library 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler,LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.ensemble import RandomForestClassifier
```

```python
# read data 
df=pd.read_csv('Data/cars.csv')
df.head()
```

```python
# Analysis Data 
df.shape
```

```python
# Checking Null 
df.isnull().sum()
```

```python
df.info()
```

```python
# Encoding Target variable
le=LabelEncoder()
df['brand']=le.fit_transform(df['brand'])
```

```python
df.info()
```

```python
# Divide the feature in independent variable and dependent variable
X=df.drop('brand',axis=1)
y=df['brand']
```

```python
# train and test data 
X_train,X_text,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape,X_text.shape,y_train.shape,y_test.shape)
```

```python
#Train the Model
dtc=DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=3,max_features=7)
dtc.fit(X_train,y_train)
```

```python
# Accuracy of train data 
y_pred_tr=dtc.predict(X_train)
print(f'Accuracy of train data is {accuracy_score(y_train,y_pred_tr)}')
```

```python
y_pred_ts=dtc.predict(X_text)
print(f'Accuracy of test data is {accuracy_score(y_test,y_pred_ts)}')
```

```python
# Selecting only the important feature 
# Dimension Reduction
features=pd.DataFrame({"Features":X_train.columns,"Importance": np.round(dtc.feature_importances_,3)})
features=features.sort_values("Importance",ascending=False)
features
```

```python
rf=RandomForestClassifier(n_estimators=100,criterion='entropy',max_depth=3)
rf.fit(X_train,y_train)
```

```python
y_pred_tr=rf.predict(X_train)
print(f'Train Accuracy is {accuracy_score(y_train,y_pred_tr)}')
```

```python
y_pred_ts=rf.predict(X_text)
print(f'Test Accuracy is {accuracy_score(y_test,y_pred_ts)}')
```

```python
# Selecting only the important feature 
# Dimension Reduction
features=pd.DataFrame({"Features":X_train.columns,"Importance": np.round(rf.feature_importances_,3)})
features=features.sort_values("Importance",ascending=False)
features
```

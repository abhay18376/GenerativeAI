```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,f1_score
```

```python
# The datset contain details of user like userid, gender, age , estimated salary, Purchase the data show where person is likely  purchased actefact advertised or not 
# we create the model to predict that person can purchased actefact advertised or not 
```

```python
# read the data 
df=pd.read_csv('Data/Social_Network_Ads.csv')
df.head()
```

```python
#Label Encoding
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df.info()
```

```python
# Scaling the input data 
sc=StandardScaler()
df[['Age','EstimatedSalary']]=sc.fit_transform(df[['Age','EstimatedSalary']])
```

```python
df.head()
```

```python
# Seperating independent and Dependent variable 
x=df[['Gender','Age','EstimatedSalary']]
y=df['Purchased']
```

```python
# train test split 
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

```python
# train the model
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
# predict 
y_pred=log.predict(X_train)
print(f'Accuracy of train {accuracy_score(y_train,y_pred)}')
```

```python
y_pred_test=log.predict(X_test)
print(f'Accuracy of test {accuracy_score(y_test,y_pred_test)}')
```

```python
cm=confusion_matrix(y_test,y_pred_test)
print(cm)
```

```python
f=f1_score(y_test,y_pred_test)
f
```

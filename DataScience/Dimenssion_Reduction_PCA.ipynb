```python
# import library 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.decomposition import PCA 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Data/Iris.csv')
df.head()
```

```python
#Standardised the data 
x=df.iloc[:,1:5]
y=df['target']
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
#Projection of PCA to 2D 
pca=PCA(n_components=2)
PrincipalComponents=pca.fit_transform(x)
PrincipalDataFrame=pd.DataFrame(data=PrincipalComponents,columns=["PC1","PC2"])
PrincipalDataFrame.head()
```

```python
# Concanating DataFrame along x axis =1
finalDF=pd.concat([PrincipalDataFrame,y],axis=1)
finalDF.head()
```

```python
# Calculating Variance Ratio 
pca.explained_variance_ratio_
```

```python
le=LabelEncoder()
df['target']=le.fit_transform(df['target'])
df.head()
```

```python
X =df.drop(columns=['target','Id'])
y=df['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
print(f'Accuracy before PCA {model.score(X_test,y_pred)*100}')
```

```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split ,cross_val_score,KFold
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
```

```python
df=pd.read_csv('Data/mtcars_for_manymerge.csv')
```

```python
df.columns
```

```python
x=df[['cyl','hp']]
y=df['Feedback']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
knn=KNeighborsClassifier(n_neighbors=4)
knn.fit(x,y)
```

```python
y_pred=knn.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
# Cross Validation Score
```

```python
knn=KNeighborsClassifier(n_neighbors=4) 
print(cross_val_score(knn,x,y,cv=7,scoring='accuracy').mean())
```

```python
log=LogisticRegression() 
print(cross_val_score(log,x,y,cv=7,scoring='accuracy').mean())
```

```python
data=pd.read_csv('Data/Diabetes.txt',sep=',',header=None)
```

```python
data.columns=['preg','plas','pres','skin','test','mass','pedi','age','class']
```

```python
data.head()
```

```python
data.columns
```

```python
X=data[['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age']]
Y=data['class']
```

```python
kfold=KFold(n_splits=10,random_state=7,shuffle=True)
model=AdaBoostClassifier(n_estimators=30,random_state=7)
result=cross_val_score(model,X,Y,cv=kfold)
print(result.mean())
```
